```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/KrisMatrix/DataScience/main/Gender_Classification/Transformed%20Data%20Set%20-%20Sheet1.csv")
df.head()
```

```python
#Before we turn all the string categories into numbers, we need to conver the 
# genders to number under a single column.

df["Gender"][df["Gender"] == "M"] = 1
df["Gender"][df["Gender"] == "F"] = -1
df.tail()
df['Gender'] = pd.to_numeric(df['Gender'])
df.dtypes
```

```python
# Turn all categories into numbers (aka one hot encoding)
data = pd.get_dummies(df)
data.head() # view the converted columns
```

```python
# Create X and y data
X = data.drop(columns=['Gender'])
y = data['Gender']
```

```python
y
```

```python
# Create training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.33, 
                                                    random_state=42) # set random state for reproducible splits
```

```python
# Set random seed
tf.random.set_seed(42)

# Create a new model (same as model_2)
model_1 = tf.keras.Sequential([
  tf.keras.layers.Dense(1),
  tf.keras.layers.Dense(1)
])

# Compile the model
model_1.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                        optimizer=tf.keras.optimizers.SGD(),
                        metrics=['mae'])

# Fit the model
model_1.fit(X_train, y_train, epochs=100)
```

```python
model_1.evaluate(X_test, y_test)
```

```python
y_pred = model_1.predict(X_test)
```

```python
# Set random seed
tf.random.set_seed(42)

# Create a new model (same as model_2)
model_2 = tf.keras.Sequential([
  tf.keras.layers.Dense(100),
  tf.keras.layers.Dense(1)
])

# Compile the model
model_2.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                        optimizer=tf.keras.optimizers.SGD(),
                        metrics=['mae'])

# Fit the model
model_2.fit(X_train, y_train, epochs=100)
```

```python
model_2.evaluate(X_test, y_test)
```

```python
# Set random seed
tf.random.set_seed(42)

# Create a new model (same as model_2)
model_3 = tf.keras.Sequential([
  tf.keras.layers.Dense(100, activation="relu"),
  tf.keras.layers.Dense(1)
])

# Compile the model
model_3.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                        optimizer=tf.keras.optimizers.SGD(),
                        metrics=['mae'])

# Fit the model
model_3.fit(X_train, y_train, epochs=100)
```

```python
# Set random seed
tf.random.set_seed(42)

# Create a new model (same as model_2)
model_4 = tf.keras.Sequential([
  tf.keras.layers.Dense(100, activation="tanh"),
  tf.keras.layers.Dense(1)
])

# Compile the model
model_4.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                        optimizer=tf.keras.optimizers.Adam(),
                        metrics=['mae'])

# Fit the model
model_4.fit(X_train, y_train, epochs=100)
```

```python
model_4.evaluate(X_test, y_test)
```
